```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Task1_data1.csv')

df.drop_duplicates()

df['Data1_concat'] = df['Order ID']+df['Product ID']
df
```

```python
df1 = pd.read_csv('Task1_data2.csv')

df1.drop_duplicates()

df1['Data2_concat'] = df1['Order ID']+df1['Product ID']
df1
```

```python
#merging based on both datasets' concatinated Order_ID,Product_ID into single data set:merged_data
merged_data = df.merge(df1, how='outer', left_on=['Data1_concat'], right_on=['Data2_concat'])
merged_data
```

# How to identify the Records (Order ID + Product ID combination) present in data1 but missing in data2 
Answer:


```python
Present_in_df_missing_in_df1 = merged_data[merged_data['Data2_concat'].isna()]
Present_in_df_missing_in_df1.shape[0]
```

# How to identify the Records (Order ID + Product ID combination) missing in data1 but present in data2
Answer:

```python
Present_in_df1_missing_in_df = merged_data[merged_data['Data1_concat'].isna()]
Present_in_df1_missing_in_df.shape[0]
```
